```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv(r"E:\projects\nature-sales-forecasting\data\dataDumpDayItemandRateWise.csv")

df['trnDate'] = pd.to_datetime(df['trnDate'], errors='coerce')

df = df[['trnDate', 'itemCategory', 'quantity']].copy()
```

```python

def category_sales(df,category):
    
    df_cat = df[df['itemCategory'] == category].copy()
    df_cat = df_cat.groupby('trnDate')['quantity'].sum().asfreq('D').fillna(0)


    plt.figure(figsize=(10,5))
    plt.title(f"Sales Trend for {category}")
    plt.plot(df_cat.values, label='Daily Sales', marker='o', linewidth=2, markersize=4, alpha=0.8)    
    plt.xlabel("Date")
    plt.ylabel("Quantity Sold")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```

```python
category_sales(df, "KHADNYA")
```
